```python
train_datagen = ImageDataGenerator(rescale=1./255, 
                                   rotation_range=10,
                                   width_shift_range=0.2,
                                   height_shift_range=0.2,
                                   shear_range=0.7,
                                   zoom_range=[0.9, 2.2],
                                   horizontal_flip=True,
                                   vertical_flip=True,
                                   fill_mode='nearest',
                                   validation_split=0.3)
```

```python
for batch in train_datagen.flow(x, batch_size=1, save_to_dir='warehouse/preview', save_prefix='tri', save_format='png'):
    i += 1
    if i > 30: 
        break
```

```python
train_generator = train_datagen.flow_from_directory('./cosmetic/',
                                                 shuffle=True, 
                                                 seed=13,
                                                 target_size = (300, 300),
                                                 batch_size = 15,
                                                 class_mode = 'categorical',
                                                 subset="training")
test_generator = train_datagen.flow_from_directory('./cosmetic/',
                                                 shuffle=True, 
                                                 seed=13,
                                                 target_size = (300, 300),
                                                 batch_size = 10,
                                                 class_mode = 'categorical',
                                                 subset="validation")
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(2, 2),
                 activation='relu',
                 input_shape=(300,300,3)))
model.add(Conv2D(64, (2, 2), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(40, activation='relu'))
model.add(Dense(23, activation='softmax'))

# 모델 엮기
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit_generator(
        train_generator,
        steps_per_epoch=15,
        epochs=10,
        validation_data=test_generator,
        validation_steps=3)
```

```python
print("-- Evaluate --")

scores = model.evaluate_generator(
            test_generator, 
            steps = 5)
```
